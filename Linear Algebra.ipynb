# Linear Algebra

```python
import numpy as np
v = np.array([2, 3, 5])
print(v)
```

```python
row_vector = np.array([1, 2, 3])  # Row vector
column_vector = np.array([[1], [2], [3]])  # Column vector
print("Row Vector:", row_vector)
print("Column Vector:\n", column_vector)
```

```python
magnitude = np.linalg.norm(v)
print("Magnitude of vector v:", magnitude)

w = np.array([4, 0, 8])
distance = np.linalg.norm(v - w)
print("Euclidean Distance between v and w:", distance)
```

```python
a = 3 #scaler
shifted_vector = v + a
print("Shifted Vector:", shifted_vector)


scaled_vector = a * v
print("Scaled Vector:", scaled_vector)


vector_sum = v + w
vector_diff = v - w
print("Vector Sum:", vector_sum)
print("Vector Difference:", vector_diff)
```

```python
dot_product = np.dot(v, w)
print("Dot Product:", dot_product)

cosine_theta = dot_product / (np.linalg.norm(v) * np.linalg.norm(w))
angle = np.arccos(cosine_theta)
print("Angle between v and w (in radians):", angle)

unit_vector = v / np.linalg.norm(v)
print("Unit Vector:", unit_vector)
```

```python
projection = (np.dot(v, w) / np.dot(w, w)) * w
print("Projection of v onto w:", projection)

```
